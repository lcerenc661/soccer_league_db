```python
soccer_col_names = ['Team','G','W','D','L','GS','GC','DIF','PTS']


def get_soccer_df(df_ligas, soccer_col_names, ligas):
    df_list = []
    for row in range(len(df_ligas)):
        df = pd.read_html(df_ligas['url'][row])
        df = pd.concat([df[0], df[1]], axis=1, ignore_index=True)
        df[0] = df[0].apply(lambda x : x[5:] if x[:2].isnumeric() else x[4:])
        df.columns = soccer_col_names
        df['League'] = ligas[row]
        df_list.append(df)
    return df_list
        
```

```python
df_list = get_soccer_df(df_ligas, soccer_col_names, ligas)
```

```python
df_list[2]['Team']
```

```python
team_df = pd.DataFrame(columns=['Team','Team_id'])
for league in df_list:
    new_df = pd.DataFrame(columns=['Team','Team_id'])
    new_df['Team'] = league['Team'].copy(deep=True)
    new_df['Team_id'] = new_df['Team'].apply(lambda x: shortuuid.uuid(x))
    new_df['Team_id'] = new_df['Team_id'].str.slice(start=0, stop=8)
    team_df = pd.concat([team_df,new_df ], axis=0, ignore_index=True)
    
team_df
```

```python
team_df.to_csv('df_teams.csv', index=False)
```
